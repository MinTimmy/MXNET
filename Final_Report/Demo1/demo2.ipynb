```python
from gluoncv import model_zoo, data, utils
from matplotlib import pyplot as plt

net = model_zoo.get_model('ssd_512_resnet50_v1_voc', pretrained=True)

# x, img_1 = data.transforms.presets.ssd.load_test('1.png', short=512)
x, img_1 = data.transforms.presets.ssd.load_test('images/street_small.jpg', short=512)
print('Shape of pre-processed image:', x.shape)

class_IDs, scores, bounding_boxs = net(x)

ax = utils.viz.plot_bbox(img_1, bounding_boxs[0], scores[0],
                         class_IDs[0], class_names=net.classes) # scores[0],
plt.show()
```
